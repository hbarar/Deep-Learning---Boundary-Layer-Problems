```python
# Python 3.6 and Tensorflow 1.14
# -----------------------------------------
# Equation: f'''+0.5*f*f"=0 ; f(0)=0 , f'(0)=0 , f'(10)=1
# ----------------------------------------
import math
import warnings
import xlsxwriter
from numpy import array
from numpy.linalg import norm

warnings.simplefilter(action='ignore', category=FutureWarning)
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io
from scipy.interpolate import griddata
from pyDOE import lhs
import time

np.random.seed(1234)
tf.set_random_seed(1234)


class PhysicsInformedNN:
    # Initialize the class
    def __init__(self, X_f, layers, lb, ub, pr):

        # X0 = np.concatenate((x0, 0 * x0), 1)  # (x0, 0)
        # X_lb = np.concatenate((0 * tb + lb[0], tb), 1)  # (lb[0], tb)
        # X_ub = np.concatenate((0 * tb + ub[0], tb), 1)  # (ub[0], tb)

        self.pr = pr
        self.lb = lb  # [-1 0]
        self.ub = ub  # [1 1]

        # self.x0 = X0[:, 0:1]  # (50,1) => x0
        self.t0 = X_f[0].reshape(1, 1)  # * np.ones((1, 1))  # X0[:, 1:2]  # (50,1)=> 0
        self.tend = X_f[-1].reshape(1, 1)
        self.t_f = X_f[1:-1, 0:1]  # allocated-y : (500,1)

        # Initialize NNs      : Initialize weight and bias as Variables
        self.layers = layers
        self.weights, self.biases = self.initialize_NN(layers)

        # tf Placeholders:
        # Everything should be defined as placeholders (boundaris and internal (x,t) and conditions (initial and boundary)
        # Note: A None value in the shape of a tensor means that the tensor can be of any size in rows

        self.t0_tf = tf.placeholder(tf.float32, shape=[None, self.t0.shape[1]])
        self.tend_tf = tf.placeholder(tf.float32, shape=[None, self.tend.shape[1]])
        self.t_f_tf = tf.placeholder(tf.float32, shape=[None, self.t_f.shape[1]])

        # tf Graphs
        # obtaining u,v and its gradients for all the boundaries and internal nodes
        self.f0_pred, self.f0_x_pred, self.f0_xx_pred, self.theta0_pred , self.theta0_x_pred  = self.net_uv(self.t0_tf)
        self.fend_pred, self.fend_x_pred, self.fend_xx_pred, self.thetaend_pred, self.thetaend_x_pred = self.net_uv(self.tend_tf)
        self.f_f_pred, self.f_theta_pred = self.net_f_uv(self.t_f_tf)

        # Loss
        self.loss = tf.reduce_mean(tf.square((0 * np.ones(1).reshape(1, 1)) - self.f0_pred)) + \
                    tf.reduce_mean(tf.square((0 * np.ones(1).reshape(1, 1)) - self.f0_x_pred)) + \
                    tf.reduce_mean(tf.square((1 * np.ones(1).reshape(1, 1)) - self.fend_x_pred)) + \
                    tf.reduce_mean(tf.square((1 * np.ones(1).reshape(1, 1)) - self.theta0_pred)) + \
                    tf.reduce_mean(tf.square((0 * np.ones(1).reshape(1, 1)) - self.thetaend_pred)) + \
                    tf.reduce_mean(tf.square(self.f_f_pred)) + tf.reduce_mean(tf.square(self.f_theta_pred))

        self.optimizer_Adam = tf.train.AdamOptimizer(learning_rate=0.0001)
        self.train_op_Adam = self.optimizer_Adam.minimize(self.loss)

        # tf session
        self.sess = tf.Session(config=tf.ConfigProto(allow_soft_placement=True,
                                                     log_device_placement=True))

        init = tf.global_variables_initializer()
        self.sess.run(init)

    def initialize_NN(self, layers):
        weights = []  # list of arrays
        biases = []  # it means weights corresponding to biases
        num_layers = len(layers)
        for l in range(0, num_layers - 1):
            W = self.xavier_init(size=[layers[l], layers[l + 1]])
            b = tf.Variable(tf.zeros([1, layers[l + 1]], dtype=tf.float32), dtype=tf.float32)
            weights.append(W)
            biases.append(b)
        return weights, biases

    def xavier_init(self, size):
        in_dim = size[0]
        out_dim = size[1]
        xavier_stddev = np.sqrt(2 / (in_dim + out_dim))
        return tf.Variable(tf.truncated_normal([in_dim, out_dim], stddev=xavier_stddev), dtype=tf.float32)

    def neural_net(self, X, weights, biases):
        num_layers = len(weights) + 1

        H = 2.0 * (X - self.lb) / (self.ub - self.lb) - 1.0  # initial scaling between -1 and 1
        for l in range(0, num_layers - 2):
            W = weights[l]
            b = biases[l]
            H = tf.tanh(tf.add(tf.matmul(H, W), b))  # Hyperbloc tangent activation function
        W = weights[-1]
        b = biases[-1]
        Y = tf.add(tf.matmul(H, W), b)  # no activation for the last layer
        return Y  # (:,2)

    def net_uv(self, x):
        X = x  # tf.concat([x, t], 1)

        uv = self.neural_net(X, self.weights, self.biases)
        f = uv[:, 0:1]
        theta = uv[:, 1:2]

        f_x = tf.gradients(f, x)[0]
        f_xx = tf.gradients(f_x, x)[0]

        theta_x = tf.gradients(theta, x)[0]

        return f, f_x, f_xx, theta , theta_x

    def net_f_uv(self, x):
        f, f_x, f_xx, theta , theta_x = self.net_uv(x)

        f_xxx = tf.gradients(f_xx, x)[0]
        theta_xx = tf.gradients(theta_x, x)[0]

        f_f = f_xxx + (0.5 * f * f_xx)
        f_theta = (theta_xx / self.pr) + (0.5 * f * theta_x)

        return f_f, f_theta

    def callback(self, loss):
        # print('Loss:', loss) ***************  Print Loss
        alaki = 1

    def train(self, nIter):

        tf_dict = {self.t0_tf: self.t0,
                   self.tend_tf: self.tend,
                   self.t_f_tf: self.t_f}

        start_time = time.time()
        workbook = xlsxwriter.Workbook('C:/Users/hasan/Desktop/loss.xlsx')
        worksheet = workbook.add_worksheet()
        roww=0
        columnn=0
        for it in range(nIter):
            self.sess.run(self.train_op_Adam, tf_dict)

            # Print
            if it % 10 == 0:
                elapsed = time.time() - start_time
                loss_value = self.sess.run(self.loss, tf_dict)
                print('It: %d, Loss: %.3e, Time: %.2f' %
                      (it, loss_value, elapsed))
                contents=[it,loss_value]

                columnn=0
                for j in range(2):
                    worksheet.write(roww, columnn, contents[j])
                    columnn+=1
                roww+=1
                start_time = time.time()
        workbook.close()
    def predict(self, X_star):  # predicting the whole domain

        # X_star is all the nodes (interior and boundaries)

        # obtain the predicted values for all the nodes
        tf_dict = {self.t0_tf: X_star}

        f_star = self.sess.run(self.f0_pred, tf_dict)
        fp_star = self.sess.run(self.f0_x_pred, tf_dict)
        fpp_star = self.sess.run(self.f0_xx_pred, tf_dict)
        theta_star = self.sess.run(self.theta0_pred, tf_dict)
        theta_x_star = self.sess.run(self.theta0_x_pred, tf_dict)


        # obtain the value of (f in the equation f=0)
        tf_dict = {self.t_f_tf: X_star}

        f_f_star = self.sess.run(self.f_f_pred, tf_dict)

        return f_star, fp_star, fpp_star, f_f_star, theta_star ,theta_x_star


if __name__ == "__main__":


    pr = 100.0
    # Domain bounds
    lb = np.array([0.0])
    ub = np.array([8.0])

    # N_b = 70  # Time condition nodes : number of points in time axis
    # N_f = 1000  # collocation points. I changed it from  10000 to 5000
    layers = [1, 300,300,300,300,300,300,  2]  # I changed 20 to 200
    #layers = [1, 200,200,2]  # I changed 20 to 200

    # =============
    # dt=(t_end-t_start)/(num_t-1)
    
    # t-end=5 ; (dt=0.05 , num_t=101)
    # t-end=5 ; (dt=0.1 , num_t=51)
    #------------------------------------
    # t-end=8 ; (dt=0.05 , num_t=161)
    # t-end=8 ; (dt=0.1 , num_t=81)
    #------------------------------------
    t_start = 0
    t_end = 8.0
    num_t = 161  # (99) devision
    t = np.linspace(t_start, t_end, num=num_t)
    t = t.reshape(t.shape[0], 1)
    # =============
    # #points on time 0 ,10
    # idx_t = np.random.choice(t.shape[0], N_b, replace=False)
    # tb = t[idx_t, :]
    # tb=t
    # collocation points
    # X_f = lb + (ub - lb) * lhs(1, N_f)
    X_f = t

    model = PhysicsInformedNN(X_f, layers, lb, ub, pr)  # model is now a object

    start_time = time.time()
    model.train(50000)  # ************ (50000)
    elapsed = time.time() - start_time
    print('Training time: %.4f' % (elapsed))

    X_star = t
    f_pred, fp_pred, fpp_pred, f_f_pred, theta_pred, theta_x_pred = model.predict(X_star)

    ######################################################################
    ############################# Writing in Excel- F & theta #######################
    ######################################################################

    workbook = xlsxwriter.Workbook('C:/Users/hasan/Desktop/saved_file.xlsx')
    worksheet = workbook.add_worksheet()

    # Start from the first cell.
    # Rows and columns are zero indexed.

    column = 0

    content = [t, f_pred, fp_pred, fpp_pred, theta_pred, theta_x_pred ]

    # iterating through content list
    for j in range(6):
        row=0
        for k in range(161):
            # write operation perform
            worksheet.write(row, column, content[j][k])

            # incrementing the value of row by one
            # with each iteratons.
            row += 1
        column+=1
    workbook.close()



    ######################################################################
    ############################# Plotting ###############################
    ######################################################################

    df = pd.read_csv('C:/Users/hasan/Desktop/Mehdi-PINN/pra-sad-infi-8-dt-0.05.csv')
    eta = df['eta']
    F_num = df['f']
    FP_num = df['fp']
    FPP_num = df['fpp']
    # -------------
    dt = pd.read_csv('C:/Users/hasan/Desktop/Mehdi-PINN/pra-sad-infi-8-dt-0.05.csv')
    eta = dt['eta']
    theta_num = dt['theta']
    thetap_num = dt['thetap']
    
    #===================================================================
    # Calculating l-2 norm relative Error:
    
    lT = norm(theta_num.to_frame()-theta_pred)/norm(theta_num)
    print ('l2-error-T:' , "{:.2e}".format(lT))
    lTP = norm(thetap_num.to_frame()-theta_x_pred)/norm(thetap_num)
    print ('l2-error-TP:' , "{:.2e}".format(lTP))
    #================================================================
    
    
    plt.plot(t, fp_pred, 'k-', linewidth=3, label="FP-pred")
    plt.plot(t, fpp_pred, 'r-', linewidth=3, label="FPP-pred")
    plt.plot(t, f_pred, 'b-', linewidth=3, label="F-pred")
    
    plt.plot(eta, FP_num, 'ko', markersize=1, label="FP-Num")
    plt.plot(eta, FPP_num, 'ro', markersize=1, label="FPP-Num")
    plt.plot(eta, F_num, 'bo', markersize=1, label="F-Num")
    
    plt.show()

    plt.plot(t, theta_pred, 'k-', linewidth=2, label="Theta-pred")
    plt.plot(eta, theta_num, 'ko', markersize=1, fillstyle="none", label="Theta-num")
    
    plt.plot(t, theta_x_pred, 'r-', linewidth=2, label="ThetaP-pred")
    plt.plot(eta, thetap_num, 'ro', markersize=1, fillstyle="none", label="ThetaP-num")

    plt.title('PINN-Flat-Plate-Temperature (Pr=500) ')
    plt.legend()
    plt.show()

    
```
